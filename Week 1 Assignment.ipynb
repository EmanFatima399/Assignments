Assignment 3: The Number Cruncher (Day 5 Loops and Conditionals)

1- Use a for loop to iterate through numbers from 1 to 20.

```python
for num in range(1, 21):  # range(1, 21) gives numbers 1 to 20
    print(num)
```

2- For each number:
 If divisible by 3, print "Fizz".
 If divisible by 5, print "Buzz".
 If divisible by both 3 and 5, print "FizzBuzz".
 Otherwise, print the number itself

```python
# Step 2: Ask the user for a number
# FizzBuzz program from 1 to 20
for num in range(1, 21):
    if num % 3 == 0 and num % 5 == 0:  # divisible by both 3 and 5
        print("FizzBuzz")
    elif num % 3 == 0:  # divisible by 3 only
        print("Fizz")
    elif num % 5 == 0:  # divisible by 5 only
        print("Buzz")
    else:  # not divisible by 3 or 5
        print(num)
```

Use a while loop to print a countdown from that number down to 1, and then print "Blast Off!".

```python
# Step 1: Iterate through numbers 1 to 20
for num in range(1, 21):
    print(num)

# Step 2: Ask the user for a number
user_number = int(input("Enter a number for countdown: "))

# Step 4: Countdown using while loop
count = user_number  # start from the user's number
while count >= 1:
    print("Countdown:", count)
    count -= 1  # decrease by 1 each iteration

print("Blast Off!")
```

Use break to stop the countdown if the user's number is greater than 10.

```python
user_number = int(input("Enter a number for countdown: "))

# Step 5: Stop countdown if number > 10
if user_number > 10:
    print("Number is greater than 10, stopping countdown.")  # break condition
else:
    count = user_number
    while count >= 1:
        print("Countdown:", count)
        count -= 1
```
